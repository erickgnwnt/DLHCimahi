```python
# Convert 'Periode/ Tanggal' to datetime
df['Periode/ Tanggal'] = pd.to_datetime(df['Periode/ Tanggal'], errors='coerce')

# Extract year from the date
df['Year'] = df['Periode/ Tanggal'].dt.year

# # Grouping by year and calculating the mean for each water quality parameter
# water_quality_trends = df.groupby('Year').mean(numeric_only=True).reset_index()

# # Displaying the aggregated water quality trends
# water_quality_trends
```

```python
df = df.rename(columns={'Lokasi/ Koordinat': 'Koordinat', 'Periode/ Tanggal': 'Periode'})
```

```python
df.isna().sum()
```

```python
df['Koordinat'].fillna(method='ffill', inplace=True)

# Let's check the updated dataframe to see if the missing values in the 'lokasi' column have been filled.
updated_missing_values = df['Koordinat'].isnull().sum()
updated_missing_values
```

```python
##droping unused column for feature extraction
##menghapus colom perhitungan pij  karena tidak digunakan untuk proses klasifikasi
df1 = df.drop(columns=df.columns[11:35]).reset_index()
df1.rename(columns={'index':'No'})
df1 = df1.drop(columns=['No'])
```

```python
##drop perhitungan nilai index fokus ke 7 parameter(tss,DO,BOD,COD,FOsfat,Fecal Coli,Total-Coliform)
##outpuut nilai Pij
df1 = df1.drop(columns=['(Ci/Lij)M2'])
```

```python
df1.isna().sum()
```

```python
##menagani missing values pada data
##DO menggunakan interpolasi linier - DO biasanya berubah secara bertahap tergantung suhu, aliran air, dan aktivitas biologis(generative AI).
##Fosfat menggunakan interpolasi linier/backward fill - Fosfat bisa berfluktuasi karena limpasan air, aktivitas manusia, dan proses biologis(generative AI).
##total-coliform  menggunakan forward fill atau rolling median - - Nilainya tidak selalu berubah secara linier, sehingga interpolasi linier bisa kurang akurat.
print(df1['DO'].std(),'Nilai Standar Deviasi DO') 
print(df1['DO'].mean(),'nILAI Rata- Rata DO')  # Melihat rata-rata
print(df['DO'].std() / df['DO'].mean(),'Nilai Koefisien Variasi')  # Koefisien variasi

print(df1['Fosfat'].std(),'Nilai Standar Deviasi Fosfat') 
print(df1['Fosfat'].mean(),'Nilai Rata- Rata Fosfat')  # Melihat rata-rata
print(df['Fosfat'].std() / df['Fosfat'].mean(),'Nilai Koefisien Variasi')  # Koefisien variasi
```

```python
##mengurutkan data ascending dari tahun ke tahun
df1 = df1.sort_values(by='Year', ascending=True)  # Mengurutkan dari tahun terkecil ke terbesar
df1 = df1.reset_index(drop=True)
```

```python
df1['DO'].interpolate(method='linear', inplace=True)  # Interpolasi linier
df1['Fosfat'].fillna(method='bfill', inplace=True) ##menggunakan metode backward fill dikarenakan 87 data diawal kosong
df1['Total-Coliform'].fillna(method='ffill',inplace=True) ##- Metode forward fill atau rolling median mungkin lebih tepat untuk menjaga pola alami.
```

```python
import numpy as np

labels = ['TSS', 'DO', 'BOD', 'COD', 'Fosfat', 'Fecal Coli', 'Total-Coliform']
stats = df_normalized[df_normalized['Year'] == 2013][labels].values.flatten()

angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False)
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, stats, 'o-', label="2015")
ax.fill(angles, stats, alpha=0.3)
ax.set_xticks(angles)
ax.set_xticklabels(labels)
plt.title('Perbandingan Parameter di Tahun 2013')
plt.show()



labels = ['TSS', 'DO', 'BOD', 'COD', 'Fosfat', 'Fecal Coli', 'Total-Coliform']
stats = df_normalized[df_normalized['Year'] == 2014][labels].values.flatten()

angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False)
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.plot(angles, stats, 'o-', label="2017")
ax.fill(angles, stats, alpha=0.3)
ax.set_xticks(angles)
ax.set_xticklabels(labels)
plt.title('Perbandingan Parameter di Tahun 2014')
plt.show()
```
